# Object Oriented Programming (OOP) in Python

This notebook explains core OOP concepts with examples.

## 1. Class and Object

```

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

s1 = Student("Pushkar", 21)
s1.display()
```

## 2. Inheritance

```

class Person:
    def greet(self):
        print("Hello!")

class Teacher(Person):
    def subject(self):
        print("I teach Python")

t = Teacher()
t.greet()
t.subject()
```

## 3. Encapsulation

```

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

acc = BankAccount(5000)
print(acc.get_balance())
```

## 4. Polymorphism

```

class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

for animal in (Dog(), Cat()):
    animal.sound()
```

## 5. Abstraction

```

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Square(Shape):
    def area(self):
        return 4 * 4

sq = Square()
print("Area:", sq.area())
```
